**Постройте корреляционную матрицу для всех числовых переменных и визуализируйте её с помощью тепловой карты:**

```python
num_cols = df.select_dtypes(include='int64').columns
num_cols = [col for col in num_cols if col not in ['CustomerID', 'Spending Score (1-100)', 'Work Experience']]

# Корреляционная матрица
correlation_matrix = df[num_cols].corr()

# Визуализация тепловой карты
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Корреляционная матрица')
plt.show()

```
